```python
d=[['<=30','high','no','fair','no'],['<=30','high','no','excellent','no'],['31...40','high','no','fair','yes'],['>40','medium','no','fair','yes'],['>40','low','yes','fair','yes'],
            ['>40','low','yes','excellent','no'],['31...40','low','yes','excellent','yes'],['<=30','medium','no','fair','no'],['<=30','low','yes','fair','yes'],['>40','medium','yes','fair','yes'],
            ['<=30','medium','yes','excellent','yes'],['31...40','medium','no','excellent','yes'],['31...40','high','yes','fair','yes'],['>40','medium','no','excellent','no']]
d = pd.DataFrame(d,columns=['age','income','student','credit_rating','buys_computer'])
d
```

```python
#A4
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
age_encoded=le.fit_transform(d['age'])
print("age_encoded:",age_encoded)
income_encoded=le.fit_transform(d['income'])
print("income:",income_encoded)
student_encoded=le.fit_transform(d['student'])
print("student:",student_encoded)
credit_rating_encoded=le.fit_transform(d['credit_rating'])
print("credit_rating:",credit_rating_encoded)
buys_computer_encoded=le.fit_transform(d['buys_computer'])
print("buys_computer:",buys_computer_encoded)
```

```python
f=zip(age_encoded,income_encoded,student_encoded,credit_rating_encoded)
features=list(f)
features
```

```python
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
model.fit(features,buys_computer_encoded)
predicted= model.predict([[0,0,0,1]])
predicted
```

```python
# This means if age<=30, income=high, credit_rating=excellent, student=no is 0. It means he cannot buy a computer.
# Similarly, all the cases can be tested using Naive Bayes Classifier
predicted1= model.predict([[0,0,0,0]])
predicted1
```
