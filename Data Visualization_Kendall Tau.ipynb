```python
df1 = df.dropna()
df1
```

```python
#correlation of all features with class feature using Kendall Tau
from scipy.stats import kendalltau
for i in df.columns:
    print(i)
    m1=df1[i]
    m2=df1['CoapplicantIncome']
    cor, _ = kendalltau(m1, m2)
    print('Kendall Tau of class with',i,': %.3f' % cor)
```
